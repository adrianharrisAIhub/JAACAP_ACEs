```python
import pyodbc
import pandas as pd
from google.colab import drive
import os
import pandas as pd
import subprocess
from io import StringIO
```

```python
filename = "SADC_2023_National.MDB"
```

```python

root_dir = "/content/"
input_path = ""  # @param {type:"string"}
filename = "SADC_2023_National.MDB"  # @param {type:"string"}
input_dir = os.path.join(root_dir, input_path)
input_file = os.path.join(input_dir, filename)
#data = pd.read_sas(input_file)

# Display maximum columns when viewing dataframes
pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", None)
```

```python
input_file
```

```python

 # Path to the .mdb file
mdb_path = input_file

# Get the list of table names using mdb-tables
table_names = subprocess.check_output(["mdb-tables", "-1", mdb_path]).decode("utf-8").splitlines()

# Print the table names
print("Table names in the .mdb file:")
for table in table_names:
    print(table)
```

```python
# Save one of the tables as a CSV (e.g., the first table)
output_filename = filename + '.csv'
output_dir = os.path.join(root_dir, input_path)
output_file = os.path.join(input_dir, output_filename)

if table_names:
    for tbl in table_names:
        if tbl == 'SADCQ':
          table_to_save = tbl
          command = ["mdb-export", mdb_path, table_to_save]
          output = subprocess.check_output(command).decode("utf-8")

    # Convert the output to a DataFrame and save as CSV

    df = pd.read_csv(StringIO(output),low_memory=False)
    df.to_csv(output_file, index=False)
    print(f"\nTable '{table_to_save}' has been saved as a CSV.")
```
